For this exercise, I need the regular expressions module we worked with in week 2. For automisation, I decided to create a function which will perform the text normalisation, so that I can then call it for each of the 2 given dictionaries. I also noticed that the removal of punctuation is equivalent to removing all non-alphanumeric characters. Because of the fact that Excercise 3 asks to return the most frequent words, I decided to do the text preprocessing now, as I assume the most frequent ones will be filler/linker words such as "and", "a" etc. which will not provide a very insightful statistic. For this, I searched stackoverflow for stop words and borrowed the method from there and included it in my function. For this, I needed to work with and import nltk, which is a module for Python that contains corpora of natural language, one of them being stop words (https://stackoverflow.com/questions/5486337/how-to-remove-stop-words-using-nltk-or-python).

```python
import re # import the module for regular expressions
import nltk # import nltk
nltk.download('stopwords') # download stopwords
from nltk.corpus import stopwords # import the stopwords corpus from the nltk package

def normalise_lyrics(albums): # define a function that takes a dictionary which contains a string of lyrics and returns a normalised version of the lyrics as output
    """
    Normalise a dictionary of song lyrics:
    1. remove punctuation (non-alphanumeric characters)
    2. convert to lowercase 
    3. remove stop words

    Args:
    - albums: a dictionary where the keys are song titles and the values are the lyrics

    Returns:
    - a new dictionary with the same keys as the input dictionary, but with normalised lyrics
    """
    stop_words = set(stopwords.words('english')) # create a set of stop words as seen on stackoverflow
    normalised_lyrics_dict = {}  # initialise an empty dictionary the function should work on
    for song_title, lyrics in albums.items(): #for loop goes through the dictionary
        lyrics = re.sub(r'\[.*?\]', '', lyrics) # remove meta-information enclosed in '[ ]'; r'\[.*?\] matches all text enclosed in square brackets, including the brackets 
        lyrics = re.sub(r'\W+', ' ', lyrics) # remove non-alphanumeric characters from the lyrics string; the r'\W+' matches one or more non-alphanumeric characters, and the second argument ' ' replaces the matched characters with a space
        lyrics = lyrics.lower() # convert to lowercase
        normalised_lyrics_dict[song_title] = lyrics # replaces the original lyrics with the normalised version for each song_title-lyrics pair in the dictionary
    normalised_lyrics_dict_final = {} # initialize an empty dictionary with the normalised lyrics of each song 
    for song_title, lyrics in normalised_lyrics_dict.items(): #for loop goes through the dictionary
        normalised_lyrics_dict_final[song_title] = ' '.join(word for word in lyrics.split() if word not in stop_words)
      # normalised_lyrics_dict[title] creates a new key-value pair in the normalised_lyrics_dict dictionary, where the key is the title of the song and the value is the normalized lyrics.
       # ' '.join(word for word in lyrics.split() if word not in stop_words) splits the lyrics into a list of words using the split() method and then filters out any stop words using the if statement. The join() method joins the remaining words back into a string with spaces between them.
    return normalised_lyrics_dict_final #return the final version of the dictionary

normalised_numan_lyrics_1979 = normalise_lyrics(numan_lyrics_1979)
normalised_numan_lyrics_2017 = normalise_lyrics(numan_lyrics_2017)

print(f'The normalised lyrics for the 1979 album: {normalised_numan_lyrics_1979}')
print(f'The normalised lyrics for the 2017 album: {normalised_numan_lyrics_2017}')

```

For this excercise I also opted for creating a function that will be called for the normalized dictionaries I got in the previous one. I broke down the function according to the questions. To calculate the "type to token ratio", I first calculate the total number of words and then the number of unique words. For the average word length I need the total length of the words and the total number of words. For the longest and shortest songs, I use if conditional statements and I go through the dictionary calculating the length of each song (by lyrics, in number of words) and comparing it to the the longest/ shortest song, updating whenever it is longer or shorter respectively. For the largest/ smallest vocabulary part I need to indentify unique words.  For the most frequent words, I used Counter, as we learnt in week 3. 

```python
from collections import Counter #imports Counter to count words as we learnt in week 3
def statistics_lyrics(lyrics_dict): #defines a function that will return the required statistics 
    """
    Analyze the lyrics of a given album and return various statistics.

    Args:
        lyrics_dict (dict): A dictionary where keys are song titles and values are song lyrics.

    Returns:
        A tuple containing the following statistics:
        - A list of the most frequent words and their counts.
        - The type to token ratio of the lyrics.
        - The average word length of the lyrics.
        - The title of the longest song (by lyrics).
        - The title of the shortest song (by lyrics).
        - A list of the song titles with the largest vocabulary.
        - A list of the song titles with the smallest vocabulary.
    """
    # calculate the total number of words and unique words
    words = [] #empty list to hold all words
    unique_words = set() #empty set to hold all unique words
    for lyrics in lyrics_dict.values(): #goes through the values of the dictionary
        lyrics_words = lyrics.split() #splits them into words
        words.extend(lyrics_words) #words are added to the 'words' list with the extend() method (week 1)
        unique_words.update(lyrics_words) #unique words are added to the 'unique_words' set witht the .update() method
    total_words = len(words) #takes the length of the 'words' list with the len() function and stores it in the variable total_worlds 
    total_unique_words = len(unique_words) #the same for the unique_words set

    # calculate the type to token ratio
    type_token_ratio = total_unique_words / total_words #calculates and stores the type to token ratio in a variable 

    # calculate the average word length
    total_word_length = sum(len(word) for word in words) #adds up the length of each string representing a word using the sum() function and stores it in a variable
    average_word_length = total_word_length / total_words #average word lenght is calculated by deviding the total word length by the number of words and it's stored in a variable

    #the longest and shortest songs (by lyrics)
    longest_song = ('', 0) #initialize a tuple, which has as first element a song title and second element the length of the lyrics in words (0 for the longest because it can only be greater than this)
    shortest_song = ('', float('inf')) #same for this tuple (infinite for the shortest because the length is always smaller than this)
    for title, lyrics in lyrics_dict.items(): #for loop goes through the dictionary
        lyrics_length = len(lyrics.split()) #lyrics are split into words with the split() method and their length is calculated with len(), and then stored in a variable
        if lyrics_length > longest_song[1]: #if the length of the current song is greater than the length of the longest_song
            longest_song = (title, lyrics_length) #update the tuple with the title of the song and the length of the lyrics
        if lyrics_length < shortest_song[1]: #if the the length of the current song is smaller than the length of the shortest_song
            shortest_song = (title, lyrics_length)#update the tuple with the title of the song and the length of the lyrics

    #the songs with the largest and smallest vocabularies
    largest_vocab_songs = [] #initializes empty list to store the titles of the largest vocabulary songs
    smallest_vocab_songs = [] #same for the titles of the smallest 
    largest_vocab_size = 0 #initialize a variable to store the size of the largest vocabulary (0 because it is always greater than this)
    smallest_vocab_size = float('inf') #same for the size of the smallest (infinite because it is always smaller than this)
    for title, lyrics in lyrics_dict.items(): #for loop that goes through the dictionary 
        vocab_size = len(set(lyrics.split())) #for each song, splits the lyrics into a list of words (split() method), removes duplicates with the set() function (because sets have only unique values in them, unlike lists), calculates the length which is equivalent to the number of unique words --> the zise of the vocabulary, stored in a variable
        if vocab_size > largest_vocab_size: #if the size of the vocabulary for the current song is greater than the largest vocabulary
            largest_vocab_songs = [title] #updates the list with the title of the current song
            largest_vocab_size = vocab_size #updates the vocabulary size to the current value (largest one so far)
        elif vocab_size == largest_vocab_size: #in the case the size is equal to the largest vocabulary size
            largest_vocab_songs.append(title) #updates the list by adding the title of the current song to it, without removing the title which previously had the largest vocabulary size
        if vocab_size < smallest_vocab_size: #same for the smallest
            smallest_vocab_songs = [title]
            smallest_vocab_size = vocab_size
        elif vocab_size == smallest_vocab_size:
            smallest_vocab_songs.append(title)

    #the most frequent words
    word_counts = Counter(words) #counts the frequency of each word in the list and stores it
    most_frequent_words = word_counts.most_common(10) #uses the most_common() method (week 3) to return a list of ther 10 most frequent words

    return (most_frequent_words, type_token_ratio, average_word_length,
            longest_song[0], shortest_song[0],
            largest_vocab_songs, smallest_vocab_songs) # return calculated values 
            
cars_dict = {"Cars": normalised_numan_lyrics_1979["Cars"]} #create a new dictionary that stores only the song "Cars" as key and its lyrics as value
stats_numan_1979 = statistics_lyrics(normalised_numan_lyrics_1979) #call the function for the first album
stats_numan_2017 = statistics_lyrics(normalised_numan_lyrics_2017) #call the function for the second album
stats_cars = statistics_lyrics (cars_dict) #call the function for the cong "Cars" (now a dictionary)

print(f"""Statistics for the 1979 album:
- The most frequent words are: {stats_numan_1979[0]}
- The type-to-token ratio is: {stats_numan_1979[1]}
- The average word length is: {stats_numan_1979[2]}
- The longest song is: {stats_numan_1979 [3]}
- The shortest song is: {stats_numan_1979 [4]}
- The song(s) with the largest vocabulary: {stats_numan_1979[5]}
- The song(s) with the smallest vocabulary: {stats_numan_1979[6]} \n""" )

print(f"""Statistics for the 2017 album:
- The most frequent words are: {stats_numan_2017[0]}
- The type-to-token ratio is: {stats_numan_2017[1]}
- The average word length is: {stats_numan_2017[2]}
- The longest song is: {stats_numan_2017[3]}
- The shortest song is: {stats_numan_2017[4]}
- The song(s) with the largest vocabulary: {stats_numan_2017[5]}
- The song(s) with the smallest vocabulary: {stats_numan_2017[6]} \n""")

print(f"""Statistics for the 'Cars' song:
- The most frequent words are: {stats_cars[0]}
- The type-to-token ratio is: {stats_cars[1]}
- The average word length is: {stats_cars[2]}
- The longest song is: {stats_cars[3]}
- The shortest song is: {stats_cars[4]}
- The song(s) with the largest vocabulary: {stats_cars[5]}
- The song(s) with the smallest vocabulary: {stats_cars[6]}""")


```
